# 3 How to rename colums :  

```python
df= pd.DataFrame({'A col':[1,2,3,4,5,6],'B col':[7,8,9,10,11,12]})
df
```

```python
df.rename(columns={'A col':'col_a','B col':'col_b'},inplace=True)
df
```

```python
df.columns=['col_1','col_2']
df
```

```python
# to replace the chacter with other character 
df.columns=df.columns.str.replace('_', '')
df
```

```python
# Adding prefix to the columns
df=df.add_prefix('baba')
df
```

```python
# Adding surfix to the columns
aqsa=df.add_suffix ('aqsu')
aqsa
```

# 8_select a colum by data_type

```python
df.dtypes  # give data type 
```

```python
# select only the numeric data
df.select_dtypes(include=['number']).head()
```

```python
# select only the numeric data and float 
df.select_dtypes(include=['float']).head()
```

```python
# select only the numeric data and float 
df.select_dtypes(include=['object','number']).head()
```

```python
# select only the numeric data and float 
df.select_dtypes(include=['object','category']).head()
```

# 9-convert strings into the numbers
