# Handling the Data Inconsistencies


### 1. INconsistent formate(e.g yyyy-mm-dd, mm_dd_yyyy, dd_mm_yyyy)
### 2. Inconsistent Naming of contries(e.g U.S.A, United_State, USA etc)
###  3. Typographical_error ( mistakes in data)
### 4. Dublication ( remove the duplicate) 
###  5. contradictory_ data ( e.g son age is less the father , some logical error occur )

```python
# import libraries 
import pandas as pd 
import numpy as np
```

```python
# Data are given : 
data = {
    'name': ['ali', '  SARA', 'JOHN ', 'Emma', 'zhang', None, 'Carlos '],
    'age': [25, 30, None, 22, 29, 35, None],
    'country': ['America ', 'Canada', 'USA', 'U.S.A', 'canada', 'spain', 'Spain'],
    'date': ['2010-01-01', '02/15/2011', '10-05-2010', '2011/08/20', None, '7-7-2010', '09-09-2011'],
    'sales_2010': [1000, 2000, None, 1500, 1800, 1300, None],
    'sales_2011': [1100, None, 2500, 1700, None, 1400, 1600]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
# fill the missing date
df['date']=df['date'].ffill()
df
```

# standarizing the date formate 

```python
df['date']= pd.to_datetime(df['date'],errors='coerce')
df['date']=df['date'].dt.strftime('%y_%m_%d')
df.head()
```

```python
df['date']=df['date'].ffill()
df.head()
```

# Harmonize the country name 


```python
country_maping = {'USA':' U.S.A', 'America': 'U.S.A', 'canada':'Canada','spain': 'Spain'}
df['country']= df['country'].replace(country_maping )
df.head()
```

# correct the typographical mistake 

```python
df['name'] = df['name'].str.strip().str.title()
df['name']= df['name'].fillna('Unknown')
```

```python
df
```

```python
df['name']=df['name'].replace({
    'ali':'Ali', 
    'zhang':'Zhang',
    'SARA':'Sara',
    'JOHN':'John'})
df
```

# Remove the contradictory data 

```python
df=df.drop(df[df['sales_2010']<=df['sales_2011']].index)
df
```
