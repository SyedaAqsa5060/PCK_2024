ğŸ”¹ What is Feature Encoding?
Feature Encoding is the process of converting categorical data (text/labels) into numerical format so that machine learning algorithms can understand and use it.

ğŸ¤” Why is it needed?
ML models like Linear Regression, Decision Trees, SVM, Neural Networks work with numbers, not text.

So if you have:

Gender	Country
Male	Pakistan
Female	USA

You can't feed this directly to a model â€” you need to encode it first.

ğŸ§° Types of Feature Encoding
Encoding Type	Use Case	How It Works
Label Encoding	Ordinal data (with order)	Assigns each category a number (e.g., Male = 0, Female = 1)
One-Hot Encoding	Nominal data (no order)	Creates separate columns for each category with 0s and 1s
Ordinal Encoding	Ordered categories (e.g., low, medium, high)	Like Label Encoding but maintains rank
Binary Encoding	High cardinality data (many categories)	Converts categories into binary numbers
Frequency / Count Encoding	Text with repeated labels	Replace each category with its count or frequency

ğŸ”¸ Examples
ğŸ“ Label Encoding
Gender	Encoded
Male	0
Female	1

ğŸ“ One-Hot Encoding
Gender_Male	Gender_Female
1	0
0	1

âœ… When to Use Which?
Situation	Use
Few categories, no order	One-Hot Encoding
Few categories, with order	Ordinal or Label Encoding
Many categories (100+)	Binary or Count Encoding

ğŸ”· Discretization kya hota hai?
Discretization ek technique hai jisme hum continuous numerical data ko categorical bins ya groups mein convert karte hain.

ğŸ”¸ Ye is liye kiya jata hai jab:

Aapko machine learning model mein categorical features chahiye hote hain.

Aap histogram plot ya category-based analysis kar rahe hote hain.

ğŸ§ª Example:
Age (Continuous)	Age Group (Discrete)

ğŸ§° Types of Discretization
Type	Description	Example
Equal Width Binning	Har bin ka range same hota hai	0â€“10, 10â€“20, 20â€“30

```python
import pandas as pd
ages = [15, 22, 37, 45, 67]
df = pd.DataFrame({'Age': ages})

# Custom Bins
bins = [0, 18, 35, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

print(df)
```
