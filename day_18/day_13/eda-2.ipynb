```python
df.drop(columns='deck', inplace=True)
```

```python
df.shape
```

```python
# visualize and explore
sns.countplot(data=df, x='survived', hue='sex')
```

```python
sns.barplot(data=df, x='survived', y='age', hue='who')
```

```python
sns.countplot(data=df, x='pclass')
```

```python
# data distribution 
sns.histplot(data=df, x='age')
```

```python
df['age'].hist(bins=10, color='red' ,alpha=0.75)
```

```python
sns.boxplot(data=df, x='sex',y='age', hue='who')
```

```python
# scatter plot 
sns.scatterplot(data=df , x='age', y='fare', hue='who')
plt.ylim(0,300) # limit of y axis min to max
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='survived')
```

```python
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
print(plt.xticks(rotation=0))
```

```python
sns.swarmplot(x='age', y='fare' ,data=df)
```

```python
sns.swarmplot(data=df, x='survived', y='age', hue='who')
```

```python
sns.catplot(data=df, x='pclass',y='fare',hue='embarked',kind='point')
```

```python
# relation of two numerical variable 
df['Fare_Category'] = df['fare'].apply(lambda x: 'Low' if x < 50 else 'High')

# Now use `Fare_Category` in `col`
sns.lmplot(data=df, x='age', y='fare', col='Fare_Category')

plt.show()
```

```python
sns.lmplot(data=df, x='age', y='fare',hue='who')
```

```python
# how add column and make the age group 
df.columns
```

```python
df['age_grouping']=pd.cut(df['age'],[0,10,18,30,50,80],
                          labels=['kids','child','youth','Middle_age','old'])
```

```python
df.columns
```

```python
df.head()
```

```python
df.groupby(df['age_grouping'])['survived'].mean().plot(kind='bar')
```

```python
sns.pairplot(df, hue='survived', vars=['age', 'fare','pclass'])
```
