## üîπ What is Regression?
Regression ek aisi supervised machine learning technique hai jo kisi cheez ki value predict karti hai ‚Äî numbers (continuous output) mein.

Example:

Ghar ki price kitni hogi?

Bill ke hisaab se tip kitni deni chahiye?

Student ki study hours ke mutabiq marks kya aayenge?

üìå Jab model se hum "kitna?" ya "kya number?" ka sawal karein ‚Üí to regression use hota hai.

üîπ Why Use Regression?
Future prediction

Business decision making

Financial forecasting

Data trend analysis

üß† Types of Regression
1. Linear Regression üìà
Sirf ek feature (input) hota hai

Input aur output ka relation straight line hota hai

Formula: y = mx + c

‚úÖ Example:
total_bill barhta hai ‚Üí tip bhi barhti hai.

2. Multiple Linear Regression
1 se zyada features hotay hain

Still straight line hoti hai but in multi-dimensional space

‚úÖ Example:
House price based on size, location, rooms

3. Polynomial Regression üîÅ
Jab relation straight nahi hota ‚Üí curve banata hai

Model higher degree polynomial banata hai:
y = a + bx + cx¬≤ + dx¬≥ + ...

‚úÖ Example:
Age vs Experience ‚Üí pehle performance barhti hai, phir ghat jaati hai

4. Ridge Regression (L2 Regularization)
Linear regression + penalty lagata hai (overfitting se bachata hai)

Extra large coefficients ko control karta hai

5. Lasso Regression (L1 Regularization)
Similar to Ridge, but yeh some features ko zero bhi kar deta hai

Useful for feature selection

6. Logistic Regression ‚ö†Ô∏è
Naam regression ka hai, lekin yeh classification task hoti hai.

Output is not number, but Yes/No, 0/1, True/False

Used for: spam detection, disease prediction, etc.

üìù Summary Table
Type	Use Case	Output Type
Linear Regression	1 feature, straight line relation	Number
Multiple Linear	Many features	Number
Polynomial Regression	Curved data	Number
Ridge / Lasso Regression	Prevent overfitting	Number
Logistic Regression	Classification (Yes/No)	0 or 1

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt  

from sklearn.preprocessing import StandardScaler  
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score  
```

```python
# load the data tips form sns
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
```

```python
# split the data into x and y 
x=df[['total_bill']]
scalar= StandardScaler()
x=scalar.fit_transform(x)
y=df['tip']
```

```python
# split the data into train test split 
x_train, x_test , y_train , y_test= train_test_split(x,y , test_size=0.2)
```

```python
# call the model 
model= LinearRegression()
```

```python
# train the model 
model.fit(x_train, y_train)
```

```python
# take a model intercept and slope, make an equation
print(model.intercept_)
print(model.coef_)
print('y=', model.intercept_,'+', model.coef_,'* x')
```

```python
# predict 
y_pred= model.predict(x_test)
```

```python
x_test
```

```python
y_pred= model.predict(x_test)
```

```python
# evluate the model 
print('MSE=',mean_squared_error(y_test, y_pred))
print('R2=',r2_score(y_test, y_pred))
print('RMSR=', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# plot the model and data
plt.scatter (x_test, y_test)
plt.plot(x_test,y_pred,color= 'red')
plt.show()
```
