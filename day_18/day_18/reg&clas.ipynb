# 💡 Train-Test Split Kya Hai?
Machine Learning me tum apna dataset do parts me todti ho:

Training set → Model ko sikhane ke liye 📚

Test set → Model ki performance check karne ke liye 📝

Example:
Jaise tum exam ke liye alag se practice questions (training) karti ho, aur phir final exam me new questions aate hain (testing).

🚨 Agar Train-Test Split Galat Ho To Kya Hota Hai
Overfitting → Model training data me expert ban jata hai, lekin new data me fail ho jata hai (jaise ratta maar ke exam me fail 😅).

Underfitting → Model itna simple hota hai ki training me bhi theek se nahi seekhta.

Data leakage → Test data ka kuch hissa training me chala jaye to results fake achhe lagte hain.

📏 Common Ratios
70/30 → 70% training, 30% testing (small data).

80/20 → Common balanced split.

90/10 → Jab data bohot zyada ho.

🎯 Kyun Matter Karta Hai
Tum test set ka result use karke decide karti ho ki model real world me kaisa chalega.

Agar split biased ho gaya, tumhare metrics (MAE, Accuracy, etc.) jhoothe achhe dikh sakte hain.

💡 Pro Tip: Data ko split karte waqt random_state set karo (Python me) taake har run me same split mile, aur test set ka data training me kabhi na aaye.
